```python
from bs4 import BeautifulSoup
import requests 
import json
import pandas as pd
import csv
from types import *
import sys

first_time = 0
biglist = []
with open('122new.csv', 'wb') as csvfile:
    writer = csv.writer(csvfile, delimiter=',',
                             quoting=csv.QUOTE_MINIMAL)

    for i in xrange(263001,270000):
        try:
            url = "https://www.chillingeffects.org/notices/{}.json".format(i)
        except:
            continue
        else:
            all_keys = []
            r = requests.get(url)
            data = r.text
            # to get the size of content in each url
            if sys.getsizeof(data)/10000 > 0:
                biglist.append(i)
                continue    
            try:
                df = pd.read_json(data)
            except:
                continue
            
            gen = 0
            for key, value in df.iteritems():
                gen += 1
                if gen == 1:
                    main_key = key
                
            #print i, "main_key: ", main_key
            
            #capture keys only first time
            if first_time == 0:
                for key,value in df[main_key].iteritems():
                    all_keys.append(key)
                    #print key,",",value
                
                writer.writerow(all_keys)
                
            all_keys = []
            #capture values        
            for key, value in df[main_key].iteritems():
                #print df['dmca'][key]
                
                if isinstance(value, StringTypes):
                    all_keys.append(value.encode('utf-8'))
                else:
                    all_keys.append(value)
                
            writer.writerow(all_keys)
            first_time = 1

print biglist
       
        #with open('chilling8.json','a') as outfile:
         #   json.dump(data,outfile)
```
